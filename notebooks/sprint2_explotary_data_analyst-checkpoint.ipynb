```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Plot Style
try:
    plt.style.use('seaborn-v0_8-darkgrid')
except:
    plt.style.use('seaborn-darkgrid')
```

```python
# Load the data
df = pd.read_csv('global_food_wastage_dataset_cleaned.csv')
```

```python
# Display transposed summary statistics
df.describe().T
```

```python
# Global food waste over time
yearly_waste = (
    df.groupby("Year")
      .sum(numeric_only=True)
      .reset_index()
      [["Year", "Total Waste (Tons)", "Economic Loss (Million $)"]]
      .sort_values("Year")
)

fig, ax1 = plt.subplots(figsize=(10, 6))
ax2 = ax1.twinx()

sns.lineplot(yearly_waste, x="Year", y="Total Waste (Tons)",
             ax=ax1, marker="o", color="tab:blue", label="Total Waste (Tons)")
sns.lineplot(yearly_waste, x="Year", y="Economic Loss (Million $)",
             ax=ax2, marker="s", color="tab:red", label="Economic Loss (Million $)")

ax1.set(title="Global Food Waste Trends 2018 - 2024",
        xlabel="Year", ylabel="Total Waste (Tons)")
ax2.set_ylabel("Economic Loss (Million $)")

ax1.legend(loc="upper left")

plt.tight_layout()
plt.show()
```

```python
# Waste by Food Category Over Time

# Group data by year and food category
category_yearly = (
    df.groupby(["Year", "Food Category"], as_index=False)
      ["Total Waste (Tons)"]
      .sum()
)

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=category_yearly,
    x="Year",
    y="Total Waste (Tons)",
    hue="Food Category",
    marker="o",
    palette="tab10"   # different color for each category
)

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Total Waste (Tons)")
plt.title("Waste Trends by Food Category")

# Legend and layout
plt.legend(title="Food Category", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
# Calculate total waste by country
country_stats = (
    df.groupby("Country", as_index=False)
      .agg({"Total Waste (Tons)": "sum"})
)

# Select top 15 countries by total waste
top_15_waste_countries = country_stats.nlargest(15, "Total Waste (Tons)")

# Plot bar chart with different colors (fixed)
plt.figure(figsize=(10, 6))
sns.barplot(
    data=top_15_waste_countries,
    y="Country",
    x="Total Waste (Tons)",
    hue="Country",        # required for palette
    palette="tab20",
    legend=False          # hide legend
)

plt.title("Top 15 Countries by Total Food Waste between 2018 -2024")
plt.xlabel("Total Waste (Tons)")
plt.ylabel("Country")
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(10, 6))

sns.regplot(
    data=df,
    x="Total Waste (Tons)",
    y="Economic Loss (Million $)",
    scatter_kws={"alpha": 0.4},
    line_kws={"color": "red"}
)

plt.title("Relationship Between Total Waste and Economic Loss")
plt.xlabel("Total Waste (Tons)")
plt.ylabel("Economic Loss (Million $)")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x="Total Waste (Tons)",
    y="Economic Loss (Million $)",
    hue="Food Category",
    alpha=0.6
)

plt.title("Economic Loss vs Total Waste by Food Category")
plt.xlabel("Total Waste (Tons)")
plt.ylabel("Economic Loss (Million $)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(11, 6))

sns.boxplot(
    data=df,
    x='Food Category',
    y='Economic Loss (Million $)'
)

plt.title('Distribution of Economic Loss by Food Category')
plt.xlabel('Food Category')
plt.ylabel('Economic Loss (Million $)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Calculate yearly totals first
yearly_totals = (
    df.groupby('Year')[['Economic Loss (Million $)', 'Total Waste (Tons)']]
      .sum()
      .reset_index()
)

# Compute cost per ton
yearly_totals['Cost per Ton ($)'] = (
    yearly_totals['Economic Loss (Million $)'] /
    yearly_totals['Total Waste (Tons)'] * 1000
)

# Plot
plt.figure(figsize=(9, 5))
plt.plot(
    yearly_totals['Year'],
    yearly_totals['Cost per Ton ($)'],
    marker='o',
    linewidth=2
)

plt.title('Average Economic Loss per Ton of Food Waste (20182024)')
plt.xlabel('Year')
plt.ylabel('Cost per Ton ($)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
corr_cols = [
    "Total Waste (Tons)",
    "Economic Loss (Million $)",
    "Avg Waste per Capita (Kg)",
    "Population (Million)",
    "Household Waste (%)"
]

corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(8, 6))
plt.imshow(corr_matrix, cmap="coolwarm", vmin=-1, vmax=1)
plt.xticks(range(len(corr_cols)), [c.replace(" (", "\n(") for c in corr_cols], fontsize=9)
plt.yticks(range(len(corr_cols)), [c.replace(" (", "\n(") for c in corr_cols], fontsize=9)
plt.title("Correlation Matrix")
plt.colorbar()

# Add correlation values on the heatmap
for i in range(len(corr_cols)):
    for j in range(len(corr_cols)):
        plt.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}", ha="center", va="center", fontsize=9)

plt.tight_layout()
plt.show()
```
